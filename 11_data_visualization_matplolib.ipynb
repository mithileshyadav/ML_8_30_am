```python
import numpy as np
import matplotlib.pyplot as plt
```

## Polynomial Regression data visualization using age and weight

```python
# create some random data
Age = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75])

Weight = np.array([5, 15, 20, 35, 80, 100, 120, 60, 45, 50, 130, 40, 35, 25, 20])
```

```python
plt.scatter( Weight, Age)
plt.xlabel("This the age.")
plt.ylabel("This the Weight")
plt.title("This the visualization of data")
plt.show()
```
