```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
# 1 Load the csv data 
pd = pd.read_csv("Data/Standardization_scaling_Pension_data.csv")
pd
```

```python
# 2 Separate the X (Feature/independent/input) 
X = pd[["Age", "Marital_status"]]
X
```

```python
# 3  y(dependent/output/target)
y = pd["Pension_Salary"]
y
```

```python
# Now i am going start Standardization(Z_Score ) sclaing
# Create a object of the StandardScaler
stnd_scaler_obj = StandardScaler()

# Now transform the data
X_scale_data = stnd_scaler_obj.fit_transform(X)
X_scale_data
```

```python
stats = {
    "mean": np.round(np.mean(X_scale_data, axis=0), 2),
    "std": np.round(np.std(X_scale_data, axis=0), 2),
}

stats

```
