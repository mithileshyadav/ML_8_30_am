```python
# imoprt all Libraries
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
# Load csv data 
df = pd.read_csv("Data/one-hot-encoding-data.csv")
df
```

```python
# Filter catergories and text data from the csv
X = df[["City", "Famous_placer"]]
X
```

```python
# create an object of OneHotEncoder classs
encoder = OneHotEncoder(sparse_output=False)
encoded_data = encoder.fit_transform(X)
encoded_data
```

```python
encoded_df = pd.DataFrame(encoded_data, columns = encoder.get_feature_names_out(X.columns))
encoded_df
```
