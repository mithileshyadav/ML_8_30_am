```python
import pandas as pd
```

```python
# load the csv file
df = pd.read_csv("Data/simple_lr_data_large_student_study_data.csv")
```

```python
# Visulize the data in matplotlib
# import matplotlib
import matplotlib.pyplot as plt

# Scattering of the data
plt.scatter(df["study_hours"], df["exam_score"])

# set the xlabel and ylabel
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")

# show the data 
plt.show()
```

```python
# split into x and y
x = df[["study_hours"]]
y = df["exam_score"]
```

```python
# laod the train_test_split
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state= 50)
```

```python
# Now i am going to import LinearRegression
from sklearn.linear_model import LinearRegression
```

```python
# Now i am going to train the Machine Learning model into LinearRegression
# creating model object of LinearRegression() class
model = LinearRegression()

# Now i am going to train the model using fit() 
model.fit(X_train, y_train)
```

```python
dir(model)
```

```python
# Now i am going to find the w/weight/coeficient
# Here is a function for find the coeficient -> coef_
my_w_coeficient = model.coef_[0]
my_w_coeficient
```

```python
# now i am going to find the Bais/intercept -> intercept_
my_bais_intercept = model.intercept_
my_bais_intercept
```
