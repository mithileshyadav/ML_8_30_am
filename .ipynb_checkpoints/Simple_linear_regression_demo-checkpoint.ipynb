```python
# ‡§Æ‡§æ‡§® ‡§≤‡•á‡§§‡•á ‡§π‡•à‡§Ç:
# X = Amount (Crore ‚Çπ), Y = Land (Bigha)
# 10 ‡§ï‡§∞‡•ã‡§°‡§º ‚Üí 1 ‡§¨‡•Ä‡§ò‡§æ, 1 ‡§ï‡§∞‡•ã‡§°‡§º ‚Üí 0.10 ‡§¨‡•Ä‡§ò‡§æ
# ‡§Ø‡§π ‡§è‡§ï ‡§∏‡•Ä‡§ß‡§æ (Linear) ‡§∞‡§ø‡§∂‡•ç‡§§‡§æ ‡§π‡•à --> ‡§á‡§∏‡§≤‡§ø‡§è ‡§Ø‡§π Simple Linear Regression ‡§π‡•à

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the csv data now
df = pd.read_csv("Data/Noida_authority_plots_Simple_linear_regression.csv")
df
```

```python
# Input data----
X = df[['Amount_in_Crore']]   # 2D required
Y = df['Land_in_Bigha']
```

```python
# Step 5: Train-Test Split (‡§¨‡§π‡•Å‡§§ ‡§ú‡§º‡§∞‡•Ç‡§∞‡•Ä)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# ‚ùì Categorical Encoding ‡§≤‡§ó‡•á‡§ó‡§æ? ‚ùå ‡§®‡§π‡•Ä‡§Ç
# ‡§ï‡•ç‡§Ø‡•ã‡§Ç‡§ï‡§ø: Amount = numeric Land = numeric
# üëâ Encoding ‡§§‡§¨ ‡§≤‡§ó‡§§‡•Ä ‡§π‡•à ‡§ú‡§¨: ==> City, Color, Gender ‡§ú‡•à‡§∏‡•á data ‡§π‡•ã‡§Ç
```

```python
# ‚ùì Feature Scaling ‡§≤‡§ó‡•á‡§ó‡•Ä? ‚ùå Simple Linear Regression ‡§Æ‡•á‡§Ç ‡§ú‡§∞‡•Ç‡§∞‡•Ä ‡§®‡§π‡•Ä‡§Ç
# ‡§ï‡•ç‡§Ø‡•ã‡§Ç? ‡§ï‡•á‡§µ‡§≤ ‡§è‡§ï feature ‡§π‡•à Values ‡§¨‡§π‡•Å‡§§ ‡§¨‡§°‡§º‡•Ä ‡§®‡§π‡•Ä‡§Ç ‡§π‡•à‡§Ç
# üëâ Scaling ‡§ú‡§∞‡•Ç‡§∞‡•Ä ‡§π‡•ã‡§§‡•Ä ‡§π‡•à:
# Gradient Descent, Multiple features, Distance based models
```

```python
#‚ùì ‡§´‡§ø‡§∞ fit_transform() ‡§ï‡§π‡§æ‡§Å use ‡§π‡•ã‡§§‡§æ ‡§π‡•à? 
# üëâ fit_transform() ‡§Ü‡§§‡§æ ‡§π‡•à:
# Preprocessing classes ‡§Æ‡•á‡§Ç
# ‡§ú‡•à‡§∏‡•á:
# StandardScaler
# MinMaxScaler
# OneHotEncoder
# PolynomialFeatures
```

```python
# ================================================================================================
# LinearRegression class ‡§Æ‡•á‡§Ç fit_transform() ‡§π‡•ã‡§§‡§æ ‡§π‡•Ä ‡§®‡§π‡•Ä‡§Ç ‡§π‡•à
# fit() ‡§ï‡•ç‡§Ø‡§æ ‡§ï‡§∞‡§§‡§æ ‡§π‡•à? (Model ‡§ï‡•á ‡§≤‡§ø‡§è)
# ‡§ú‡§¨ ‡§π‡§Æ LinearRegression().fit() ‡§ï‡§∞‡§§‡•á ‡§π‡•à‡§Ç: too Model: Data ‡§ï‡•ã ‡§¶‡•á‡§ñ‡§§‡§æ ‡§π‡•à --> Relationship ‡§∏‡•Ä‡§ñ‡§§‡§æ ‡§π‡•à --> Best m (slope) ‡§î‡§∞ c (intercept) ‡§®‡§ø‡§ï‡§æ‡§≤‡§§‡§æ ‡§π‡•à
# ‡§Ø‡§π‡•Ä actual model training ‡§π‡•à -->  ‡§¨‡§ø‡§®‡§æ fit ‡§ï‡•á prediction ‡§®‡§π‡•Ä‡§Ç ‡§π‡•ã ‡§∏‡§ï‡§§‡§æ
```

```python
# transform() ‡§ï‡•ç‡§Ø‡§æ ‡§ï‡§∞‡§§‡§æ ‡§π‡•à?
# üëâ Data ‡§ï‡•ã ‡§¨‡§¶‡§≤‡§§‡§æ ‡§π‡•à,
# ‡§≤‡•á‡§ï‡§ø‡§® ‡§ï‡•Å‡§õ ‡§®‡§Ø‡§æ ‡§∏‡•Ä‡§ñ‡§§‡§æ ‡§®‡§π‡•Ä‡§Ç‡•§
# X_scaled = scaler.transform(X)
```

```python
# fit_transform() ‡§ï‡•ç‡§Ø‡§æ ‡§ï‡§∞‡§§‡§æ ‡§π‡•à?
# üëâ ‡§Ø‡§π ‡§¶‡•ã ‡§ï‡§æ‡§Æ ‡§è‡§ï ‡§∏‡§æ‡§• ‡§ï‡§∞‡§§‡§æ ‡§π‡•à: ==> fit() + transform()
# ‚úî ‡§™‡§π‡§≤‡•á data ‡§∏‡•á mean & std ‡§∏‡•Ä‡§ñ‡§§‡§æ ‡§π‡•à
# ‚úî ‡§´‡§ø‡§∞ ‡§â‡§∏‡•Ä data ‡§ï‡•ã scale ‡§ï‡§∞‡§§‡§æ ‡§π‡•à
```

```python
# Model Training going to start
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Prediction
Y_pred = model.predict(X_test)
Y_pred
```

```python
# Y_pred = model.predict(X_test)
# Why Prediction mein Prediction X_test he kyo dete hain aur kuch kyo nhi like X_train/y_test/y_train?
# üëâ ‡§ï‡•ç‡§Ø‡•ã‡§Ç‡§ï‡§ø ‡§π‡§Æ model ‡§ï‡•ã ‚Äú‡§®‡§è / ‡§Ö‡§®‡§¶‡•á‡§ñ‡•á data‚Äù ‡§™‡§∞ test ‡§ï‡§∞‡§®‡§æ ‡§ö‡§æ‡§π‡§§‡•á ‡§π‡•à‡§Ç
# üëâ ML ‡§ï‡§æ ‡§Ö‡§∏‡§≤‡•Ä ‡§Æ‡§ï‡§∏‡§¶ ‡§Ø‡§π‡•Ä ‡§π‡•à
```

```python
# Predict the model for 10 Crore so output will be 1
model.predict([[10]])
```

```python
# Predict the model for 100 Crore so output will be 10
model.predict([[100]])
```

```python
# Hamara model train ho gya. Abb hum ishe Website k sath integrate kar sakte hain
# Uuse se pahle Check karte hai accuracy
```

```python
# Model Evaluation (Error Check)
# Mean Squared Error
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, Y_pred)
print("MSE:", mse)
# ‚úî ‡§¨‡§π‡•Å‡§§ ‡§ï‡§Æ MSE = ‡§Ö‡§ö‡•ç‡§õ‡§æ model
```

```python
# R¬≤ Score (‡§∏‡§¨‡§∏‡•á Important)
from sklearn.metrics import r2_score
r2 = r2_score(y_test, Y_pred)
print("R2 Score:", r2)
```

```python
# Interpretation:
# R¬≤ = 1 ‚Üí Perfect fit
# R¬≤ > 0.9 ‚Üí Excellent
# R¬≤ < 0.5 ‚Üí Bad model
```

```python
# Underfitting / Overfitting ‡§ï‡•à‡§∏‡•á ‡§™‡§π‡§ö‡§æ‡§®‡•á‡§Ç?
# üî¥ Underfitting
# Train score ‡§ï‡§Æ
# Test score ‡§ï‡§Æ
# üëâ Model ‡§¨‡§π‡•Å‡§§ simple
```

```python
# üî¥ Overfitting
# Train score ‡§¨‡§π‡•Å‡§§ high
# Test score low
# üëâ Model data ‡§Ø‡§æ‡§¶ ‡§ï‡§∞ ‡§∞‡§π‡§æ ‡§π‡•à
```

```python
# üü¢ Good Fit (‡§π‡§Æ‡§æ‡§∞‡§æ case)
# Train score ‚âà Test score
# R¬≤ ‚âà 1
# ‚úî ‡§ï‡•ç‡§Ø‡•ã‡§Ç‡§ï‡§ø data perfect linear ‡§π‡•à
```
