# Ridge Regression (L2 Regulariztion)

```python
import numpy as np
```

```python
# creating X input feature data (Independent variable)
X = np.array([[1, 10], [1.5, 15], [2, 20], [2.5, 25], [3, 30], [3.5, 35], [4, 40] ])
# creating target (Y) values (Depenedent variable)
Y = np.array([200, 250, 300, 350, 400, 450, 500])
```

```python
# Now i am going for train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3)
```

#### Linear regression without Regulariztion(L2 Ridge)

```python
from sklearn.linear_model import LinearRegression
# now going to create an object
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

##### get the co-efficient(slope(w)) and Intercept(b or bias)

```python
print("Linear Regresstion co-efficient(slope(w)): ", lr.coef_) # it will provide you 2 -> co-efficient
print("Linear Regresstion Intercept(b or bias): ", lr.intercept_)
```

###### Find the Train and test Score

```python
print("Linear Regresstion Train Score: ", lr.score(X_train, y_train)) # Train Score will print
print("Linear Regresstion Test Score ", lr.score(X_test, y_test))     # Test Score will print
```

# Now i am going to add Penlaty using L2 Regularization (Ridge Regression)

```python
# importing the Ridge 
from sklearn.linear_model import Ridge
# creating an object of Ridge class
ridge_obj = Ridge(alpha = 1)
ridge_obj.fit(X_train, y_train)
```

##### Finding the score of Ridge

```python
print("Ridge Regresstion Train Score: ", ridge_obj.score(X_train, y_train)) # Train Score will print
print("Ridge Regresstion Test Score: ", ridge_obj.score(X_test, y_test))     # Test Score will print
```

# Now Visulization of both Linear Regression and Ridge Regression

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], Y, color = "blue", label = "Size vs Price")
plt.scatter(X[ : , 1], Y, color = "green", label = "Age vs Price")
# finding the prediction
lr_pred = lr.predict(X)
ridge_pred = ridge_obj.predict(X)

# Now ploting for both Linear and Ridge
plt.plot(X[ : , 0], lr_pred, color="red", label = "Linear Regression")
plt.plot(X[ : , 0], ridge_pred, color="orange", label = "Ridge Regression")

plt.xlabel("Features")
plt.ylabel("Price")
```
