# Linear regression vs Polynomial regression

```python
import numpy as np
import matplotlib.pyplot as plt
# my data for Linear regression and Polynomial regression
X = np.array([[1], [2], [3], [4], [5], [6], [7]])
y = np.array([45000, 50000, 60000, 80000, 110000, 150000, 200000])

# Normal data visualition
# data visualiztion
plt.scatter(X, y, color = "blue")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

## Implementing Linear Regression

```python
from sklearn.linear_model import LinearRegression
# creating a object of LinearRegression
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
```

## Visualiztion of Linear Regression

```python
plt.scatter(X, y, color = "blue")
plt.plot(X, y_pred, color = "orange")
plt.title("X vs Y")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

## Implemeting polynomial

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
# creating an object of PolynomialFeatures
poly = PolynomialFeatures(degree=2)

x_poly = poly.fit_transform(X)
x_poly
```

```python
# creating an object of LinearRegression
lr_model = LinearRegression()
lr_model.fit(x_poly, y)

lr_y_pred = lr_model.predict(x_poly)
```

## Visualiztion of Polynomial Regression

```python
plt.scatter(X, y, color = "blue")
plt.plot(X, lr_y_pred, color = "orange")
plt.title ("X vs Y in Polynomial regresssion")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
